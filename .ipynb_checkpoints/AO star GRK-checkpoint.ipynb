```python
class Node:
    def __init__(self, index,cost, parent=None, visited=False,
                 is_solved=False,and_map=False, or_map=False):
        self.index=index
        self.cost=cost
        self.visited=visited
        self.is_solved=is_solved
        self.and_map= and_map
        self.or_map = or_map
        self.children=()
        self.bestmove=None
        self.parent=None

    def __str__(self):
        return f'{self.index}: {self.cost}'

    def set_children(self,ch):
        self.children=ch
```

```python
#Use case 1
adj=[]
n_nodes = 21
#heuristic costs
cost=[None,0,40,2,4,1,2,3,50,60,70,80,4,5,8,9,6,7,90,90,90,90]
and_edges={}
for i in range(n_nodes+1):
  n=Node(i, cost[i])
  adj.append(n)

adj[1].set_children((adj[2],adj[3],adj[4]))
adj[2].set_children((adj[5],adj[6],adj[7]))
adj[3].set_children((adj[8],adj[9]))
adj[4].set_children((adj[10],adj[11]))
adj[5].set_children((adj[12],adj[13]));
adj[6].set_children((adj[14],adj[15]))
adj[7].set_children((adj[16],adj[17]));
adj[8].set_children((adj[18],))
adj[9].set_children((adj[19],));
adj[10].set_children((adj[20],));
adj[11].set_children((adj[21],))

and_edges[adj[1]] = (adj[3],adj[4])
adj[3].and_map = adj[4].and_map = True

and_edges[adj[2]] = (adj[5],adj[6], adj[7])
adj[5].and_map = adj[6].and_map = adj[7].and_map=True

for a in adj:
    if len(a.children)==0: a.is_solved=True
    if a.and_map==False:
       a.or_map=True
    print(f'{a.index} and {a.and_map} or {a.or_map}')
```

```python
#Use case 2
adj=[]
n_nodes = 10
#heuristic costs
cost=[None,0,6,12,9,5,7,1,4,4,1]
and_edges={}
for i in range(n_nodes+1):
    n=Node(i, cost[i])
    adj.append(n)

adj[1].set_children((adj[2],adj[3],adj[4]))
adj[2].set_children((adj[5],adj[6]))
adj[3].set_children((adj[7],))
adj[4].set_children((adj[8],adj[9]))
adj[5].set_children((adj[10],));

and_edges[adj[1]] = (adj[2],adj[3])
adj[2].and_map = adj[3].and_map = True

and_edges[adj[4]] = (adj[8],adj[9])
adj[8].and_map = adj[9].and_map =True

for a in adj:
    if len(a.children)==0: a.is_solved=True
    if a.and_map==False: a.or_map=True
    print(f'{a.index} and {a.and_map} or {a.or_map}')



for idx, ae in and_edges.items():
    print('\n',idx.index)
    for c in ae:
        print(c.index, end= " ")
```

```python
#Check whether a node is in and-nodes.
#If so return parent node and nodes in the and_arc
def get_key(and_edges, c):
    for idx, ae in and_edges.items():
        if c in ae:
            return idx, ae
```

```python
# Function updates the cost and solved status of a node.
# Changes are propagted upwards by recursively calling the function
def update_node(node):
    print("Updating node ", node.index)
    old_cost = node.cost
    old_solved = node.is_solved
    temp_cost = MAX; temp_map={}
    #Check the cost and solved status of or and and arcs
    for c in node.children:
        if temp_map.get(c,False): continue;

        if c.and_map: # if the child is in the and arc
            temp_solved=True
           #calculate the cost and check if there are more nodes in the and edge
            idx,ae = get_key(and_edges,c)
            cc=0
            for aek in ae:
                cc+=aek.cost+EDGE
                temp_map[aek]=True
                temp_solved=temp_solved and aek.is_solved
            if temp_solved:
                node.is_solved=True

        else: # else if child is in the or edge
            cc = c.cost+EDGE
            temp_map[c]=True
            if (c.is_solved):
                node.is_solved=True
            #print(f' **or edge {c.index} {temp_cost}')
        # if node is solved, further checking is stopped
        if (node.is_solved):
            temp_cost = cc
            node.bestmove=c
            break
        elif (cc < temp_cost):
            temp_cost= cc
            node.bestmove=c
#head is explored now update the best value of head i.e. temp_cost
    node.cost=temp_cost
    print(f'Updated node cost {node.cost} is_solved {node.is_solved}')
    # if the cost or solved status of the node is changed,
    # then propagate upwards unless it is the root node
    if (old_cost != node.cost or old_solved != node.is_solved):
        if (node != adj[1]):
            update_node(node.parent)
```

```python
#Search for the next node to expand by following best path sign
def expand_next(head):
    node = head
    while (node.visited):
        child = node.bestmove
        if child.or_map:
            node = child
        else:
            idx,ae = get_key(and_edges,child) # get nodes in and arc
            for aek in ae:
                if (not aek.is_solved):  # move to unsolved in and arc
                    node = aek
                    break
    print("Expanding node:", node.index)
    node.visited=True
    for child in node.children:
        child.parent=node
    update_node(node)
```

```python
def aostar(head):
    iter = 0
    while head.is_solved==False and iter <MAX and head.cost < MAX:
        print(f'\n  **Iteration {iter}')
        expand_next(head)
        iter+=1

    print('\n Updated costs')
    for a in adj:
        print(a.index,': ',a.cost, a.is_solved)
```

```python
MAX=1000
EDGE=1 #g cost of edge, change to 1 for use case 2
aostar(adj[1])
```
